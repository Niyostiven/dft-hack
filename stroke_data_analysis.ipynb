# 📊 Types of Data Analysis: What, How, and When

---

## 1. Descriptive Analysis
**What it is:**  
Summarizes and describes features of a dataset (e.g. averages, counts, distributions).

**Key techniques:**  
- Mean, median, mode
- Frequency tables
- Charts (bar, pie, histogram)

**When to use:**  
- At the beginning of any analysis
- To understand the structure and basic trends in data

---

## 2. Exploratory Data Analysis (EDA)
**What it is:**  
Deeper exploration of the data to find patterns, outliers, relationships, and anomalies.

**Key techniques:**  
- Univariate, bivariate, multivariate visualizations
- Correlation matrices
- Boxplots, scatter plots
- Missing value handling
- Outlier detection

**When to use:**  
- Before building models
- To prepare and clean data
- To decide what features might be important

---

## 3. Inferential Analysis
**What it is:**  
Uses a sample to make generalizations about a population.

**Key techniques:**  
- Hypothesis testing (e.g., t-tests, chi-square)
- Confidence intervals
- P-values and significance

**When to use:**  
- When making statistical conclusions
- In research to validate findings with limited data

---

## 4. Predictive Analysis
**What it is:**  
Uses historical data to predict future outcomes using machine learning or statistical models.

**Key techniques:**  
- Regression
- Classification
- Decision trees, random forest, logistic regression

**When to use:**  
- When you want to forecast or predict (e.g., who might get stroke, customer churn)

---

## 5. Diagnostic Analysis
**What it is:**  
Finds the root cause of a problem or trend observed in descriptive or exploratory analysis.

**Key techniques:**  
- Drill-down analysis
- Correlation and cause-effect investigation

**When to use:**  
- When something unexpected is observed (e.g., sudden rise in stroke cases)

---

## 6. Prescriptive Analysis
**What it is:**  
Suggests actions or decisions based on predictive insights.

**Key techniques:**  
- Optimization
- Recommendation systems
- Scenario testing

**When to use:**  
- For decision-making (e.g., what intervention reduces stroke risk the most?)

---

## 7. Causal Analysis
**What it is:**  
Tries to establish cause-and-effect relationships (e.g., Does smoking cause stroke?)

**Key techniques:**  
- Randomized controlled trials
- A/B testing
- Quasi-experimental design

**When to use:**  
- In scientific research or impact evaluation studies

---





# 📊 Types of Analysis by Number of Variables

---

## 1. Univariate Analysis
**What it is:**  
Analysis of a single variable to understand its distribution, central tendency, or spread.

**Goal:**  
- Describe the variable (e.g., how many patients fall in each age group?)
- Detect outliers
- Summarize characteristics

**Example:**
- Analyze `age` of patients.
  - Mean age
  - Histogram of age
  - Boxplot of BMI

**Code Example:**
```python
sns.histplot(data=stroke_data, x='age', kde=True)


## 2. Bivariate Analysis

**What it is:**  
Analysis of two variables to find relationships or differences between them.

**Goal:**  
- Compare one variable against another  
- Detect correlation or influence

**Examples:**  
- Compare `stroke` and `gender`  
  → Does stroke occur more in males or females?  

- Compare `age` vs `bmi`  
  → Do older patients have higher BMI?

**Code Examples:**
```python
# Categorical vs Categorical
sns.countplot(data=stroke_data, x='gender', hue='stroke')

# Numerical vs Numerical
sns.scatterplot(data=stroke_data, x='age', y='bmi')


## 3. Multivariate Analysis

**What it is:**  
Analysis involving three or more variables at the same time.

**Goal:**  
- Understand complex relationships  
- Identify combined effects  
- Prepare for machine learning

**Examples:**  
- Analyze how `age`, `bmi`, and `hypertension` relate to `stroke`  
- Examine if `smoking_status` and `residence_type` together affect stroke risk

Skewness interpretation:

- **~0 (around 0)** → Symmetric distribution (like a bell curve)
- **> 0 (positive)** → Right-skewed (tail to the right, common with high outliers)
- **< 0 (negative)** → Left-skewed (tail to the left)

We plotted the histogram for `bmi` to help decide how to impute the missing values (~4% missing). The histogram shows the **distribution** (shape) of `bmi`, and this guides the imputation method.

### How histogram helps decide imputation:

- **Mean imputation**  
  Use this when the data is **normally distributed** (symmetric bell shape).  
  ❌ Not ideal here because the `bmi` histogram is **right-skewed**.

- **Median imputation**  
  Best when the data is **skewed** or has **outliers**, since the median is not affected by extreme values.  
  ✅ This is the **best choice** for `bmi` in our case.

- **Mode imputation**  
  Only used when the feature is **categorical** or has repeated values.  
  ❌ Not suitable for `bmi`, since it's continuous.

- **Advanced methods (e.g., KNN imputation)**  
  These use other features to predict missing values.  
  ✅ Possible, but more complex and not always necessary for small missing percentages like this.

**Compute IQR and detect outliers for each continuous variable**

````python
def outlier_detector(df,column):
  Q1=df[column].quantile(0.25)
  Q3=df[column].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  outliers=df[(df[column]<lower_bound) | (df[column]>upper_bound)]
  return {
      'column':column,
      'Q1':Q1,
      'Q3':Q3,
      'Lower bound':lower_bound,
      'Upper bound':upper_bound,
      'outlier count':len(outliers),
      'outlier percentage':round(len(outliers)/len(df)*100,2)
  }
````

````python
outlier_detector(stroke_df,'age')
````

````python
output=[
    outlier_detector(stroke_df,'age'),
    outlier_detector(stroke_df,'avg_glucose_level'),
    outlier_detector(stroke_df,'bmi')
]
````

````python
outlier_summary=pd.DataFrame(output)
````

````python
outlier_summary
````

The table shows the results of outlier detection using the IQR method for age, average glucose level, and BMI. The IQR method calculates lower and upper bounds to define outliers, but it does not consider whether those bounds make sense in real life. For example, the lower bound for age is -29 years, which is impossible because age cannot be negative. This highlights that while the IQR method is mathematically correct, its results must be interpreted with common sense and knowledge of the subjec

````python
def generate_ftable(df,column):
  freq=df[column].value_counts().sort_index()
  percent=(freq/len(df)*100).round(2)
  dict={
      'Frequency':freq,
      'Percentage (%)':percent
  }
  table=pd.DataFrame(dict)
  return table
````

The Chi-square test of independence is used to determine whether two categorical variables are related or independent of each other. At its core, the test compares what we observe in our data with what we would expect to see if there were no relationship between the variables. It helps answer questions like, “Does stroke occurrence differ between men and women?” or “Is smoking status related to disease presence?

# Chi2 (Chi-square statistic)
The Chi-square statistic is a number that reflects how different the observed data is from what we would expect if the two variables were not related. Conceptually, it measures the size of the “gap” between what actually happened and what would happen in a world where the variables are completely independent. A larger Chi2 value means the differences between observed and expected values are bigger, which could be a sign of a real association between the variables. If the value is small, it means the data closely match what would be expected by chance, suggesting no relationship.

# p-value
The p-value answers this question: “If there really were no relationship between these variables, what is the probability that we’d see differences this big just by luck?” A low p-value (usually less than 0.05) means that the differences are unlikely to be random and therefore may reflect a real association. A high p-value means the differences could easily occur by chance, so we conclude there is no evidence of a meaningful relationship between the variables. The p-value is the most commonly used number in deciding whether the result is statistically significant.
